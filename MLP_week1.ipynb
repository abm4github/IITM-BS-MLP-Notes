## Why do df.columns and df.shape has no '()' while df.info() and df.head() do?

Short explanation:

* df.columns, df.shape â†’ values (attributes)
* df.info(), df.head() â†’ actions (methods)

Further explanation:

* df.info()        # does something (runs)
* df.head()        # does something

* df.columns       # shows stored data
* df.shape         # shows stored data


One-line rule:
* () = action, no () = value

## Selecting desired values

### using loc(), example: loc[0:2, 'A':'C'] it considers the rows 0,1,2 (row 2 is inclusive) and the columns A, B, C (column C is inclusive). loc takes ranges in the form of names/ labels.

```python
df_random.loc[0:2, 'A':'C']  # the range of rows, the range of columns
```

###
using iloc(), example: loc[0:3, 0:3] it considers the rows 0,1,2 (row 3 is NOT inclusive) and the columns 0,1,2 (column 3 is NOT inclusive). iloc takes the ranges in the form of integers

```python
df_random.iloc[0:3, 0:3]
```

## Using lambda function.

```python
""" Here we are using the lambda function to take the dataframe as input,
using iloc, it provides the output such that the rows are in the range 0 to 3; that is, 0, 1, 2, and similarly, the columns are in the range 0 to 3; that is, 0, 1, 2. """
selector_function = lambda df: df.iloc[0:3, 0:3]
selector_function(df_random)
```

using & - bitwise AND operator

```python
""" Here we are using lambda function on df_random instead of providing the ranges directly.
I'm using a lambda function where the lambda takes the data frame and filters
columns as follows:
data frame's column A must have values greater than 0.3
data frame's column B must have a greater than 0.3 """

selector = lambda df: (df['A'] > 0.3) & (df['B'] > 0.3)
df_random[selector]
```

## Aggregation and Grouping

What it means:

* Grouping â†’ split data into groups

* Aggregation â†’ calculate something for each group (sum, mean, count, etc.)

In short:
* Split â†’ Apply â†’ Combine

```python
# creating dataframe

data = {
    'Department': ['IT', 'IT', 'HR', 'HR', 'Sales'],
    'Salary': [50000, 60000, 40000, 45000, 70000]
}

df = pd.DataFrame(data)
df
```

Grouping (groupby)

```python
df.groupby('Department') # it only splits data, no calculation yet
```

Aggregation (mean, sum, count)

```python
df.groupby('Department')['Salary'].mean() # gives average salary of each department
```

```python
df.groupby('Department')['Salary'].sum() # gives the sum of salaries in each department
```

```python
df.groupby("Department")['Salary'].count() # gives the number of employes in each department
```

Multiple aggregations (agg). Let us combine these three aggregations

```python
df.groupby('Department')['Salary'].agg(['mean','sum','count'])
```

### Grouping and Aggregating with multiple columns

```python
import pandas as pd

data = {
    'Department': ['IT','IT','IT','HR','HR','HR','Sales','Sales'],
    'Gender': ['Male','Female','Male','Male','Female','Female','Male','Female'],
    'Salary': [50000, 65000, 70000, 42000, 48000, 40000, 72000, 38000]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby(['Department', 'Gender'])['Salary'].agg(['mean', 'sum', 'count'])
# groupby is initiated with Department then inside each Department, groupby is done based on Gender
```
